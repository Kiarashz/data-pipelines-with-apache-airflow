```python
import json
import xml.etree.ElementTree as ET
from bs4 import BeautifulSoup
```

```python
with open('result2.txt', 'r') as file:
  html_page = file.read()
```

```python
print(html_page)
```

```python
root = ET.fromstring(html_page)
for child in root:
    print(child.tag, child.attrib, child.text)
```

```python
soup = BeautifulSoup(html_page, 'html.parser')

type(soup.find_all('script')[0])
# print(soup.find_all('script')[0])
# raw = json.loads(soup.find_all('script')[0])
# print(raw)
```

```python
from lxml import html

tree = html.fromstring(html_page)
title = tree.xpath('//title/text()')
script = tree.xpath('//script/text()')
print(title[0])  # Output: The Title
```

```python
import re

print(type(script[5]))
matched = re.search(r'numberSets.*?divisions', str(script[5]))
```

```python
cleaned = str(matched.group()).replace(',"divisions', '').replace('numberSets":', '')
print(cleaned)
```

```python
jc = json.loads(cleaned)
print(jc[0].get('numbers'))
win_numbers = [num.get('number') for num in jc[0].get('numbers')]
print(win_numbers)
```
